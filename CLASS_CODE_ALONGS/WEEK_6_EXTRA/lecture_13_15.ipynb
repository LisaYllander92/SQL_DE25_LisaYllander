## Lecture 13 - Subquery
- a query inside of another query
- gets a temporary result set, which can be used

```python
avg = query_database(
    """--sql
SELECT 
    AVG(grade_math)
FROM staging.students;
"""
)
```

find all rows where score is higher than the average

- subquery inside paranthesis - evalauated first

```python
query_database(
    """--sql
    SELECT 
        *
    FROM staging.students
    WHERE grade_math > (    
        SELECT 
            AVG(grade_math)
        FROM staging.students
        );
""")
```

solve without subquery - by hard coding (using the result from before)

```python
query_database(
    """--sql
    SELECT 
        *
    FROM staging.students
    WHERE grade_math > 84.2 
""")  
```

```python
# using f-sting to put in query
query_database(f"""--sql
    SELECT
        *
    FROM 
        staging.students
    WHERE grade_math > {avg['avg(grade_math)'].values[0]}
""")
```

## Lecture 14 - Views
- a saved SQL query as a virtual table
- the query is executed everytime it is accessed
- no data are stored

Purposes of Views

- restrict accesses to underlying tables
- simplify complex queries (for examples, involving joins of many tables) for users
- customerize data for different users
- maintain stable interface between underlying tables and downstream consumers (for example, when column names change in the underlying tables)

```python
query_database(
    """--sql
    FROM staging.housing_sales;
""").head()
```

pick out average of property types and put into a view to give to downstrem user
- if storing this as a table and the underlying data changes e.g. more houses are added
then the "average sales table" is wrong

- with a view - it will run teh sql query when the downstream user is quering the view - which will give fresh results all the time


```python
query_database(
    """--sql
CREATE OR REPLACE VIEW staging.average_price_type AS 
SELECT
    property_type,
    AVG(sale_price_usd) AS avg_price
FROM staging.housing_sales
 GROUP BY property_type
 """)
```
